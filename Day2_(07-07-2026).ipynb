1. Dictionary

A dictionary is a collection of key-value pairs. It is mutable, unordered (Python 3.7+ preserves insertion order), and keys must be unique.

```python
# Dictionary Example

student = {
    "Name": "Jayvina",
    "Age": 20,
    "Branch": "Computer"
}

print("Original Dictionary:", student)

# 1. get()
print("Name:", student.get("Name"))

# 2. keys()
print("Keys:", student.keys())

# 3. values()
print("Values:", student.values())

# 4. update()
student.update({"City": "Surat"})
print("Update:", student)

# 5. pop()
student.pop("Age")
print("Pop:", student)
```

2. List

A list is an ordered collection of items. It is mutable, allows duplicate values, and can store different data types.

```python
# List Example

fruits = ["Apple", "Banana", "Mango"]

print("Original List:", fruits)

# 1. append()
fruits.append("Orange")
print("Append:", fruits)

# 2. insert()
fruits.insert(1, "Grapes")
print("Insert:", fruits)

# 3. remove()
fruits.remove("Banana")
print("Remove:", fruits)

# 4. pop()
fruits.pop()
print("Pop:", fruits)

# 5. sort()
fruits.sort()
print("Sort:", fruits)
```

3. Tuple

A tuple is an ordered collection of items. It is immutable, allows duplicate values, and cannot be modified after creation.

```python
# Tuple Example

numbers = (10, 20, 30, 20, 40)

print("Original Tuple:", numbers)

# 1. count()
print("Count of 20:", numbers.count(20))

# 2. index()
print("Index of 30:", numbers.index(30))

# 3. len()
print("Length:", len(numbers))

# 4. max()
print("Maximum:", max(numbers))

# 5. min()
print("Minimum:", min(numbers))
```

4. Set

A set is an unordered collection of unique items. It is mutable and does not allow duplicate values.

```python
# Set Example

colors = {"Red", "Blue", "Green"}

print("Original Set:", colors)

# 1. add()
colors.add("Yellow")
print("Add:", colors)

# 2. remove()
colors.remove("Blue")
print("Remove:", colors)

# 3. discard()
colors.discard("Black")
print("Discard:", colors)

# 4. pop()
removed = colors.pop()
print("Removed Element:", removed)
print("Pop:", colors)

# 5. clear()
new_set = {1, 2, 3}
new_set.clear()
print("Clear:", new_set)
```

5. if Statement

The if statement is used to execute a block of code when a condition is true.

```python
# if Example

age = 20

if age >= 18:
    print("You are eligible to vote.")
```

6. if-else Statement

The if-else statement executes one block if the condition is true and another block if it is false.

```python
# if-else Example

number = 7

if number % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

7. if-elif-else Statement

The if-elif-else statement checks multiple conditions and executes the matching block.

```python
# if-elif-else Example

marks = 78

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
else:
    print("Grade C")
```

8. Nested if-else

A nested if statement contains another if statement inside it.

```python
# Nested if Example

age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Entry Allowed")
    else:
        print("Bring your ID")
else:
    print("Not Eligible")
```

9. Break Statement

The break statement is used to stop a loop immediately.

```python
# Break Example

for i in range(1, 6):
    if i == 4:
        break
    print(i)
```

10. Continue Statement

The continue statement skips the current iteration and moves to the next iteration.

```python
# Continue Example

for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

11. Pass Statement

The pass statement does nothing. It is used as a placeholder for future code.

```python
# Pass Example

for i in range(1, 4):
    if i == 2:
        pass
    print(i)
```

12. Input Method

The input() function is used to take input from the user.

```python
# Input Example

name = input("Enter your name: ")

print("Welcome", name)
```

13. range() Function

The range() function generates a sequence of numbers.

```python
# range() Example

for i in range(1, 6):
    print(i)
```

14. len() Function

The len() function returns the number of items in an object.

```python
# len() Example

city = "Rajkot"

print("Length:", len(city))
```

15. type() Function

The type() function returns the data type of a variable.

```python
# type() Example

number = 25

print(type(number))
```

16. For Loop

A for loop is used to iterate over a sequence such as a list, tuple, or string.

```python
# For Loop Example

fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
```

17. While Loop

A while loop executes a block of code as long as the condition is true.

```python
# While Loop Example

count = 1

while count <= 5:
    print(count)
    count += 1
```
